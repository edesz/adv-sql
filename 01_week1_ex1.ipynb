#### Step 1. Create a query in Snowflake that returns all customers that can place an order with Virtual Kitchen. If the customer is able to order from us, then their city/state will be present in our database. Create a query in Snowflake that returns all customers that can place an order with Virtual Kitchen.

Use the following approach

1. clean customer address city and state name columns
2. clean US cities city and state name columns
3. perform `LEFT JOIN` using 1. and 2.

````python
%%time
query = f"""
        -- 1. clean customer address table columns
        WITH t1 AS (
            -- select cleaned customer address columns
            SELECT customer_id,
                   customer_city,
                   customer_state,
                   RTRIM(LTRIM(LOWER(customer_city), ' '), ' ') AS customer_city_cleaned,
                   RTRIM(LTRIM(UPPER(customer_state), ' '), ' ') AS customer_state_cleaned
            FROM customers.customer_address
        ),
        -- 2. clean us_cities table columns
        t2 AS (
            SELECT city_name,
                   state_abbr,
                   RTRIM(LTRIM(LOWER(city_name), ' '), ' ') AS city_name_cleaned,
                   RTRIM(LTRIM(UPPER(state_abbr), ' '), ' ') AS state_abbr_cleaned,
                   geo_location
            FROM resources.us_cities
        ),
        -- 3. perform LEFT JOIN using cleaned customer address and us_cities
        -- tables
        t3 AS (
            SELECT t1.customer_id,
                   cd.first_name,
                   cd.last_name,
                   t1.customer_city,
                   t1.customer_state,
                   t2.geo_location
            FROM t1
            LEFT JOIN t2 ON t1.customer_city_cleaned = t2.city_name_cleaned
            AND t1.customer_state_cleaned = t2.state_abbr_cleaned
            LEFT JOIN (
                SELECT customer_id,
                       first_name,
                       last_name
                FROM customers.customer_data
            ) AS cd USING (customer_id)
        )
        SELECT *
        FROM t3
        """
_ = show_sql_df(query, cur, conn, True, True)
````

#### Step 2. Determine which supplier is closest to each customer, and how far the shipment needs to travel to reach the customer. Determine which supplier is closest to each customer, and how far the shipment needs to travel to reach the customer.

Use the following approach

1. in the US cities table
   - (step 1) clean city and state name columns 
   - (step 2) get single city ID per city (i.e. without duplicated city IDs)
   - (step 3) filter `us_cities` table to only capture cities without duplicated city ID
2. in the customer address table
   - (step 4) clean city and state name columns
3. (step 5) perform `INNER JOIN` using cleaned outputs of 1. and 2. above
   - the `INNER JOIN` selects customers whose city/state is in the the database and these customers are eligible to place an order from Virtual Kitchen
4. (step 6) perform `LEFT JOIN` using supplier info and cleaned `us_cities` table on the city name (lowercase) and state name (uppercase) columns
5. (step 7) perform the following to get the distance between each customer and every supplier
   - `CROSS JOIN` using customer (step 5) and supplier info (step 6) outputs from above
   - calculate distance between all resulting customer-supplier combinations
   - for each customer, assign a rank column based on the distance where the closest distance is assigned the highest rang (rank 1)
7. (step 8) get the closest supplier to each customer and order the result by last name and then by first name
   - filter the distances for all customer-supplier combinations to get the rank 1 combinations only

````python
%%time
query = f"""
        -- exercise 1 part 1
        -- - clean the city and state columns and then remove duplicated city IDs from the us_cities table
        -- - clean the city and state columns in customer_address table
        -- - INNER JOIN the two cleaned outputs from above, to get eligible customers
        --   (INNER JOIN and not LEFT JOIN is used since we only want customers whose city/state is in the database)
        -- exercise 1 part 2
        -- - LEFT JOIN supplier info and cleaned us_cities tables
        -- - CROSS JOIN the eligible customers with the supplier info to get combinations of customers and suppliers, calculate distance for all combinations, assign ranks per customer to the distance column
        -- - select the rank 1 rows, to get the closest supplier to each customer
        --
        -- EXERCISE 1 PART 1
        -- 1. clean us_cities table columns by removing whitespaces,
        -- changing city to lowercase and keeping the state abbreviation
        -- in uppercase
        WITH t1 AS (
            SELECT city_id,
                   city_name,
                   state_abbr,
                   -- remove whitespace in city name
                   RTRIM(LTRIM(LOWER(city_name), ' '), ' ') AS city_name_cleaned,
                   -- remove whitespace in state abbreviation
                   RTRIM(LTRIM(UPPER(state_abbr), ' '), ' ') AS state_abbr_cleaned,
                   geo_location
            FROM resources.us_cities
        ),
        -- 2. from cleaned us_cities columns, get cities with the first city ID
        t2 AS (
            SELECT city_name,
                   state_abbr,
                   city_name_cleaned,
                   state_abbr_cleaned,
                   MIN(city_id) AS city_id
            FROM t1
            GROUP BY ALL
        ),
        -- 3. filter original us_cities table to only capture cities with
        -- unique city ID in cleaned us_cities columns
        t3 AS (
            SELECT t2.city_id,
                   t2.city_name,
                   t2.state_abbr,
                   t1.city_name_cleaned,
                   t1.state_abbr_cleaned,
                   t1.geo_location
            FROM t1
            INNER JOIN t2 USING (city_id, city_name_cleaned, state_abbr_cleaned)
        ),
        -- 4. clean customer address table columns by removing whitespaces,
        -- changing city to lowercase and keeping the state abbreviation
        -- in uppercase
        t4 AS (
            SELECT customer_id,
                   customer_city,
                   customer_state,
                   -- remove whitespace in city name
                   RTRIM(LTRIM(LOWER(customer_city), ' '), ' ') AS customer_city_cleaned,
                   -- remove whitespace in state abbreviation
                   RTRIM(LTRIM(UPPER(customer_state), ' '), ' ') AS customer_state_cleaned
            FROM customers.customer_address
        ),
        -- 5. perform INNER JOIN using cleaned customer address and us_cities
        -- tables to get geo_location for each customer in the cleaned customer
        -- address table. The `INNER JOIN` selects customers whose city/state
        -- is in the the database and these customers are eligible to place
        -- an order from Virtual Kitchen. Perform the INNER JOIN using the
        -- cleaned columns (columns with the _cleaned suffix). However, in the
        -- output, exclude columns with the _cleaned suffix. Finally, perform
        -- a LEFT JOIN between the the output and the customers_data table in
        -- order to capture the customer's first and last name.
        t5 AS (
            SELECT t4.customer_id,
                   t4.customer_state,
                   cd.first_name,
                   cd.last_name,
                   cd.email,
                   t3.geo_location
            FROM t4
            -- JOIN with cleaned and non-duplicated us_cities table
            INNER JOIN t3 ON t4.customer_city_cleaned = t3.city_name_cleaned
            AND t4.customer_state_cleaned = t3.state_abbr_cleaned
            -- JOIN with customer_data table to get first_name and last_name
            LEFT JOIN (
                SELECT customer_id,
                       first_name,
                       last_name,
                       email
                FROM customers.customer_data
            ) AS cd USING (customer_id)
        ),
        -- EXERCISE 1 PART 2
        -- 6. perform LEFT JOIN using supplier info and cleaned us_cities
        -- tables to get geo_location for each supplier in the supplier
        -- info table. Perform the JOIN using the cleaned version of the city
        -- and state columns. However, in the output, keep the original columns
        -- from the tables.
        t6 AS (
            SELECT si.supplier_id,
                   si.supplier_name,
                   si.supplier_city,
                   t3.geo_location
            FROM suppliers.supplier_info AS si
            LEFT JOIN t3 ON LOWER(si.supplier_city) = LOWER(t3.city_name)
            AND si.supplier_state = UPPER(t3.state_abbr)
        ),
        -- 7. get cartesian product between (every possible combination of)
        -- customers and suppliers, calculate distance between all
        -- combinations of customers and suppliers, and assign rank based on
        -- distance where the smallest distance for each customer is assigned
        -- a rank of 1
        t7 AS (
            SELECT customer_id,
                   first_name,
                   last_name,
                   email,
                   supplier_id,
                   supplier_name,
                   -- get distance in km between each supplier and customer
                   st_distance(t6.geo_location, t5.geo_location) / 1000 AS distance_km,
                   -- assign rank per customer based on distance to all
                   -- suppliers, where smallest distance has the highest
                   -- rank (rank 1)
                   RANK() OVER(PARTITION BY customer_id ORDER BY distance_km) AS rank_distance_km
            FROM t5
            -- get cartesian product 
            CROSS JOIN t6
        ),
        -- 8. get smallest customer-supplier combination (closest supplier to
        -- each customer) by selecting rows with a rank of 1, which
        -- corresponds to the smallest distance between the each customer and
        -- a supplier, and sort by customer's last name and then by first name
        t8 AS (
            SELECT * EXCLUDE(distance_km, rank_distance_km),
                   distance_km AS shipping_distance_km
            FROM t7
            WHERE rank_distance_km = 1
            ORDER BY last_name, first_name
        )
        SELECT *
        FROM t8
        """
_ = show_sql_df(query, cur, conn, True, True)
````
